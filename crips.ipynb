```python
from pandas_datareader import data, wb
import os 
import pandas as pd
import numpy as np
from datetime import datetime
%matplotlib inline
```

```python
start = datetime(2016,1,1)
end = datetime(2021,2,18)
```

```python
BTC = data.DataReader('BTC-USD', 'yahoo', start, end)
BTC.head()
```

```python
ETH = data.DataReader('ETH-USD', 'yahoo', start, end)
ETH.head()
```

```python
LTC = data.DataReader('LTC-USD', 'yahoo', start, end)
LTC.head()
```

```python
XRP = data.DataReader('XRP-USD', 'yahoo', start, end)
XRP.head()
```

```python
cryptos = ['BTC', 'ETH', 'LTC', 'XRP']

cryptos.sort()

cryptos
```

```python
crips = pd.concat([BTC, ETH, LTC, XRP], keys = cryptos, axis=1)
crept = pd.concat([ETH, LTC], keys = cryptos, axis=1)
```

```python
crips
```

```python
crips.columns.names = ['CRYPTOS', 'INFO']
crept.columns.names = ['CRYPTOS', 'INFO']

crept.head()
```

MAX CLOSE PRICE

```python
crips.xs(key='Close',axis=1, level='INFO').max()
```

```python
returns = pd.DataFrame()
```

```python
for crip in cryptos:
    returns[crip + ' Return'] = crips[crip]["Close"].pct_change()

returns
```
